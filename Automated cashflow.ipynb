```python
### Programme to calculate the vendor and customer credit terms and proportions

#getting the sales

cashflow = {}

cashflow['months'] = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
cashflow['number_days'] = [31, 28,31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
cashflow['sales_forecast'] = []
cashflow['sales_cost_forecast'] = []

counter = 0

print('Please, put numbers only')
while counter < 12:
    try:
        sale_forecast =  int(input("Enter the sales forecast: "))
        cashflow['sales_forecast'].append(sale_forecast)
        sale_cost_forecast = int(input("Enter the sales cost forecast: "))
        cashflow['sales_cost_forecast'].append(sale_cost_forecast)
        counter += 1
        continue
    except:
        print("Put in a valid format")
```

```python
print(cashflow)
```

```python
import pandas as pd
import numpy as np

cashflow = pd.DataFrame(cashflow, columns = ['months', 'number_days','sales_forecast','sales_cost_forecast'])
cashflow.set_index('sales_forecast')

#Considering the Customer Credit Term, though by me as the ammount of time the company may take to receive the full amount
#of money from the sales

cashflow['forecasted_sales_considering_customer_credit_term'] = np.where(cashflow['number_days'] <= 30,
                                           cashflow['sales_forecast'] * 0.3,
                                           cashflow['sales_forecast'])

cashflow['to_be_paid_next_month'] = np.where(cashflow['number_days'] <= 30,
                                           cashflow['sales_forecast'] * 0.4,
                                           cashflow['sales_forecast'])

cashflow['to_be_paid_in_two_months'] = np.where(cashflow['number_days'] <= 30,
                                           cashflow['sales_forecast'] * 0.3,
                                           cashflow['sales_forecast'])

cashflow['to_be_paid_next_month'] = np.where(cashflow['number_days'] == 31,
                                           cashflow['sales_forecast'] * 0,
                                           cashflow['sales_forecast'])

cashflow['to_be_paid_in_two_months'] = np.where(cashflow['number_days'] >= 31,
                                           cashflow['sales_forecast'] * 0.3,
                                           cashflow['sales_forecast'])

cashflow
```

```python
cashflow['to_be_paid_this_month'] = cashflow['to_be_paid_next_month'].shift(periods=1, fill_value=0) + cashflow['to_be_paid_next_month'].shift(periods=2, fill_value=0)
cashflow
```

```python
cashflow['updated_new_sales_forecast'] = cashflow['to_be_paid_this_month'] + cashflow['forecasted_sales_considering_customer_credit_term']
cashflow
```

```python
#Considering the Vendor Credit Term, though by me as the ammount of time the company will have to pay for the whole ammount
#it has purchase from third parties

cashflow['forecasted_sales_cost_considering_vendor_credit_term'] = np.where(cashflow['number_days'] <= 30,
                                           cashflow['sales_cost_forecast'] * 0.5,
                                           cashflow['sales_cost_forecast'])

cashflow['to_be_paid_next_month(vendor)'] = np.where(cashflow['number_days'] <= 30,
                                           cashflow['sales_cost_forecast'] * 0.5,
                                           cashflow['sales_cost_forecast'])

cashflow['to_be_paid_next_month(vendor)'] = np.where(cashflow['number_days'] == 31,
                                           cashflow['sales_cost_forecast'] * 0,
                                           cashflow['sales_cost_forecast'])

cashflow
```

```python
cashflow['to_be_paid_this_month(vendor)'] = cashflow['to_be_paid_next_month'].shift(periods=1, fill_value=0)
cashflow
```

```python
cashflow['updated_new_sales_cost_forecast'] = cashflow['to_be_paid_this_month(vendor)'] + cashflow['forecasted_sales_cost_considering_vendor_credit_term']
cashflow
```

```python
updated_cashflow = cashflow.drop(columns=['sales_forecast','sales_cost_forecast'])
updated_cashflow
```

```python
updated_cashflow = cashflow.drop(columns=['sales_forecast','sales_cost_forecast', 'number_days','to_be_paid_next_month',
                                         'to_be_paid_in_two_months','to_be_paid_next_month'])
updated_cashflow
```

```python
updated_cashflow['updated_cashflow'] = updated_cashflow['updated_new_sales_forecast'] - updated_cashflow['updated_new_sales_cost_forecast']

updated_cashflow.set_index('months')
```
